### Fill missing values

```python
# for train_df
train_df["Delivery_person_Age"] = pd.to_numeric(train_df["Delivery_person_Age"], errors="coerce")
train_df["Delivery_person_Ratings"] = pd.to_numeric(train_df["Delivery_person_Ratings"], errors="coerce")
train_df["multiple_deliveries"] = pd.to_numeric(train_df["multiple_deliveries"], errors="coerce")
```

```python
# for test data

test_df["Delivery_person_Age"] = pd.to_numeric(test_df["Delivery_person_Age"], errors="coerce")
test_df["Delivery_person_Ratings"] = pd.to_numeric(test_df["Delivery_person_Ratings"], errors="coerce")
test_df["multiple_deliveries"] = pd.to_numeric(test_df["multiple_deliveries"], errors="coerce")
```

```python
train_df["Delivery_person_Age"].fillna(train_df["Delivery_person_Age"].median(), inplace=True)
train_df["Delivery_person_Ratings"].fillna(train_df["Delivery_person_Ratings"].median(), inplace=True)
train_df["multiple_deliveries"].fillna(train_df["multiple_deliveries"].median(), inplace=True)

test_df["Delivery_person_Age"].fillna(test_df["Delivery_person_Age"].median(), inplace=True)
test_df["Delivery_person_Ratings"].fillna(test_df["Delivery_person_Ratings"].median(), inplace=True)
test_df["multiple_deliveries"].fillna(test_df["multiple_deliveries"].median(), inplace=True)
```

```python
# categorical columns to mode

categorical_columns = ["Road_traffic_density", "Festival", "City"]

for col in categorical_columns:
    train_df[col].fillna(train_df[col].mode()[0], inplace=True)
    test_df[col].fillna(test_df[col].mode()[0], inplace=True)
```
